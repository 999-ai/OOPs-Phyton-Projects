```python
class Employee:
  def __init__(self,name,salary,department):
     self.name=name
     self.salary=salary
     self.department=department
  
  def show_info(self):
     print(f"name={self.name},salary={self.salary},department={self.department}")

class Manager(Employee):
  def __init__(self,name,salary,department,team_member):
       super().__init__(name,salary,department)
       self.team_member=team_member
   
  def show_info(self):
       super().show_info()
       print(f"team member:{self.team_member}")

class Executive(Manager):
  def __init__(self,name,salary,department,team_member,budget_management):
     super().__init__(name,salary,department,team_member)
     self.budget_management=budget_management

  def show_info(self):
       super().show_info()
       print(f"budget management:{self.budget_management}")


employee1=Employee("Sibel",3000,"Finance")
employee1.show_info()


manager1=Manager("Ali",5000,"Human resources",2)
manager1.show_info()


executive1=Executive("Ayse",2000,"Ä±T",3,"Budgetmanagement")
executive1.show_info()
```

```python
class Order:
  def __init__(self,order_name,quantity,price):
   self.order_name=order_name
   self.quantity=quantity
   self.price=price

  def calculate_order(self):
      return self.quantity*self.price

class Customer:
  def __init__(self,name):
   self.name=name
   self.order_list=[]

  def add_order(self,order):
      self.order_list.append(order)
      print(f"{order.order_name} is added into order list")

  def cancel_order(self,unwanted_order_name):
       self.order_list = [order for order in self.order_list if order.order_name != unwanted_order_name.order_name]
      
       print(f"{unwanted_order_name.order_name} is cancelled")
  def calculate_total_cost(self):
     return sum(order.calculate_order() for order in self.order_list)
  
  def display_all(self):
      print(f"Order list belonging to {self.name}")
      for order in self.order_list:
       print(f"order:{order.order_name},quantity:{order.quantity},price:{order.price}")


customer1=Customer("Nick")
order1=Order("Kebap",5,7)
order2=Order("Ayran",2,3)

customer1.add_order(order1)
customer1.add_order(order2)
customer1.display_all()

print(f"total cost:{customer1.calculate_total_cost()}")

customer1.cancel_order(order1)
customer1.display_all()


print(f"total cost:{customer1.calculate_total_cost()}")
```

```python
class Bank_Account:
    def __init__(self,name,bank_balance):
        self.name=name
        self.bank_balance=bank_balance
    def make_deposit(self,deposit):
        self.bank_balance= self.bank_balance+deposit
        print(f" {deposit} is made as deposit .Bank balance belonging to {self.name} is {self.bank_balance}")
    def withdrawing_deposit(self,wanted_money):
        if self.bank_balance>=wanted_money:
           self.bank_balance= self.bank_balance-wanted_money
           print(f"{wanted_money} is withdrawn.Bank balance belonging to {self.name} is {self.bank_balance}")
        else:
            print("insufficient balance")
    def display_all_deposit(self):
         print(f"Bank balance belonging to {self.name} is {self.bank_balance}")

bank1=Bank_Account("jerry",1000)
bank1.make_deposit(100)
print()
bank1.withdrawing_deposit(200)
print()
bank1.display_all_deposit()



            
```

```python
class Product:
     def __init__(self,name,price):
        self.name=name
        self.price=price
     def show_info(self):
        print(f"name:{self.name} price:{self.price}")
class Basket:
     def __init__(self):
        self.shopping_list=[]
     def add_product(self,product):
        self.shopping_list.append(product)
        print(f"{product.name} is added into the shopping_list")
     def call_off_product(self,unwanted_product):
        self.shopping_list=[product for product in self.shopping_list if product.name!=unwanted_product.name]
        print(f"{unwanted_product.name} is cancelled in the shopping list")
     def calculate(self):
        return sum(product.price for product in self.shopping_list)
       
     def display_all(self):
         if self.shopping_list:
             print("in your shopping_list your products:")
             for product in self.shopping_list:
                 product.show_info()
         else:
              print("your shopping_list is empty")


product1=Product("apple",7)
product2=Product("kiwi",9)
basket1=Basket()

basket1.add_product(product1)
basket1.add_product(product2)
basket1.display_all()
print(f"Total cost is {basket1.calculate()} TL")
print()


basket1.call_off_product(product1)
basket1.display_all()
print(f"total cost is {basket1.calculate()}TL")                             
```

```python
class Weather:
     def __init__(self,city,temperature,atmospheric_condition):
      self.city=city
      self.temperature=temperature
      self.atmospheric_condition=atmospheric_condition
     def show_info(self):
         print(f"city:{self.city},temperature:{self.temperature},atmospheric condition:{self.atmospheric_condition}")

class Weather_Forecast:
     def __init__(self):
        self.weather_forecast_list=[]
     def add_weather_forecast(self,weather):
        self.weather_forecast_list.append(weather)
        print(f"{weather.city} is added into weather condition list")
     def display_all(self):
         if self.weather_forecast_list:
            print("The weather forecast is below")
            for weather in self.weather_forecast_list:
                weather.show_info()
         else:
             print("Weather forecast list is empty")

weather1=Weather("New York",20,"sunny")
weather2=Weather("Sydney",15,"cloudy")
weather3=Weather("Dusseldorf",18,"snowy")
weather4=Weather("London",19,"rainy")

weather_forecast1=Weather_Forecast()
weather_forecast1.add_weather_forecast(weather1)
weather_forecast1.add_weather_forecast(weather2)         
weather_forecast1.add_weather_forecast(weather3)   
weather_forecast1.add_weather_forecast(weather4)
print()
weather_forecast1.display_all()
```

```python
class Songs:
    def __init__(self,song,singer,year):
     self.song=song
     self.singer=singer
     self.year=year
    def info(self):
     print(f"song:{self.song},singer:{self.singer},year:{self.year}")

class Playlist:
     def __init__(self):
         self.playlist=[]
     def add_songs(self,song):
         self.playlist.append(song)
         print(f"{song.song} is added into playlist")
     def cancel_song(self,unwanted_song):
         self.playlist=[song for song in self.playlist if song.song!=unwanted_song.song]
         print(f"{unwanted_song.song} was put out from playlist")
     def display_all(self):
         if self.playlist:
             print("the playlist is here")
             for song in self.playlist:
                 song.info()

song1=Songs("Shape of my heart","Stings",1993)
song2=Songs("Du hast mich","Ramstein",1997)
song3=Songs("Sexy ladies","Beyonce",2008)       
         
playlist1=Playlist()
playlist1.add_songs(song1)
playlist1.add_songs(song2)
playlist1.add_songs(song3)
print()
playlist1.display_all()
print()
playlist1.cancel_song(song2)
print()
playlist1.display_all()
```

```python
class Films:
    def __init__(self,name,manager,year):
     self.name=name
     self.manager=manager
     self.year=year
    def info(self):
     print(f"the name of film:{self.name},manager:{self.manager},year:{self.year}")

class Film_archive:
    def __init__(self):
     self.film_archive_list=[]
    def add_film(self,wanted_film):
      self.film_archive_list.append(wanted_film)
      print(f"{wanted_film.name} is added into film archive list")
    def call_off(self,unwanted_film):
      self.film_archive_list=[film for film in self.film_archive_list if film.name!=unwanted_film.name]  
      print(f"{unwanted_film.name} was put out from film archive list")
    def display_all(self):
      if  self.film_archive_list:
          for film in self.film_archive_list:
              film.info()
      else:
          print("film archive list ist empty")

class Film_Watched:
    def __init__(self):
        self.film_list_watched=[]
    def add_film(self,film_watched):
      self.film_list_watched.append(film_watched) 
      print(f"{film_watched.name} is added into film list watched")
    def call_off(self,unwanted_film):
      self.film_list_watched=[film for film in self.film_list_watched if film.name!=unwanted_film.name]  
      print(f"{unwanted_film.name} was put out from film list watched ")
    def display_all(self):
      
      if  self.film_list_watched:
          print("You can see the film list watched") 
          for film in self.film_list_watched:
              film.info()
      else:
          print("film list watched ist empty")

film1=Films("Barbie","Greta Gerwig",2023)
film2=Films("Gladiator","Ridley Scott",2000)
film3=Films("Kung Fu Panda","Tai Lung",2024)
film_archive=Film_archive()
film_watched=Film_Watched()
print()
film_archive.add_film(film1)
film_archive.add_film(film2)
film_archive.add_film(film3)
print()
film_archive.display_all()
print()
film_archive.call_off(film2)
print()
film_archive.display_all()
print()

film_watched.add_film(film1)
film_watched.add_film(film2)
print()
film_watched.display_all()
print()
film_watched.call_off(film2)
print()
film_watched.display_all()
```
