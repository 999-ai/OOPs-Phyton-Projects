```python
class Book:
  def __init__(self,title,writer): 
      self.title=title
      self.writer=writer
      self.availability=True

  def borrowing_book(self):
      if   self.availability:
           self.availability=False
           print(f"{self.title} was given to user")
           
      else:
           
           print(f"{self.title} has already given to user")
  def returning_book(self):
       if  not self.availability:
            self.availability=True
            print( f"{self.title} is returned to library")

       else:
            print(f"{self.title} is not borrowed and returned")
class Library:
  def __init__(self):
      self.book_list=[]
  def add_book(self,book):
      self.book_list.append(book)
      print(f"{book.title} is added into book list")
  def display_all(self):
      print("Book list is here")
      for book in self.book_list:
         position= "Available" if book.availability else "Not available" 
         print(f"book title:{book.title} writer:{book.writer} position:{position}")
  def finding_book(self,wanted_book):
       for book in self.book_list:
         if book.title==wanted_book.title:
            index= self.book_list.index(wanted_book)
            print(f"{wanted_book.title} is found and its index is {index}")
            return book
             
       return None

library1=Library()
book1=Book("Crime and Punishment","Dotsyoveski")
book2=Book("Sefiller","Dotsyoveski")

library1.add_book(book1)
library1.add_book(book2)
print()
library1.display_all()
print()
library1.finding_book(book1)
print()
book1.borrowing_book()
library1.display_all()
book1.returning_book()
print()
library1.display_all()
```
